# Stocks and Bonds

Stocks:-
•	Stocks is something that person invest money into and gains share in that company or institution
•	Stocks get dividend but no guarantee depending upon the company.
•	Voting rights available in the company.
•	Stocks can be in form of equity and mutual funds which are subjected to market risk.

Bonds:-
•	Bonds is something that person tends loan to company or institution and get interest
•	Interest is guaranteed
•	Preferential treatment given when bonds mature.
•	Bonds like Government and Corporate Bonds.

# Step-3 : Moving Windows

```python
daily_close_ambj=ambj[['Adj Close']] #Adj Close 12-02-2022-Adj Close 11-02-2022

#Daily returns
daily_pct_change_ambj=daily_close_ambj.pct_change()

#Replace NA values with 0
daily_pct_change_ambj.fillna(0,inplace=True)
daily_pct_change_ambj.head()
```

```python
daily_pct_change_ambj.hist(bins=50)
plt.show()
```

```python
daily_close_ambj=ambj[['Adj Close']]
daily_pct_change_ambj=daily_close_ambj.pct_change()
daily_pct_change_ambj.fillna(0,inplace=True)
daily_pct_change_ambj.head()

daily_close_knst=knst[['Adj Close']]
daily_pct_change_knst=daily_close_knst.pct_change()
daily_pct_change_knst.fillna(0,inplace=True)
daily_pct_change_knst.head()

daily_close_niti=niti[['Adj Close']]
daily_pct_change_niti=daily_close_niti.pct_change()
daily_pct_change_niti.fillna(0,inplace=True)
daily_pct_change_niti.head()

daily_close_hpct=hpct[['Adj Close']]
daily_pct_change_hpct=daily_close_hpct.pct_change()
daily_pct_change_hpct.fillna(0,inplace=True)
daily_pct_change_hpct.head()


```

```python
import seaborn as sns
sns.set()
```

```python
import seaborn as sns
#Set up the mataplotlib figure
f,axes=plt.subplots(2,2,figsize=(12,7))

#Plot a simple histogram with binsize determined automatically
sns.distplot(daily_pct_change_ambj['Adj Close'],color="b",ax=axes[0,0], axlabel='AMBUJA CEMENTS')

#Plot a kernel density estimate and rug plot
sns.distplot(daily_pct_change_knst['Adj Close'],color="r",ax=axes[0,1], axlabel='KUNSTOFF INDUSTRIES')

#Plot a filled kernel density estimate
sns.distplot(daily_pct_change_niti['Adj Close'],color="g",ax=axes[1,0], axlabel='NITIN SPINNERS')

#Plot a histogram and kernel density estimate
sns.distplot(daily_pct_change_hpct['Adj Close'],color="m",ax=axes[1,1], axlabel='HP COTTONS')
```

# Step 4 : Volatility

```python
import numpy as np
```

```python
min_periods = 75

#Calculate the volatality
vol=daily_pct_change_ambj.rolling(min_periods).std()*np.sqrt(min_periods)
vol.fillna(0,inplace=True)
vol.tail()
```

```python
#Plot the volatality

vol.plot(figsize=(10,8))

#Show the plot
plt.show()
```

# Rolling Means (Trends and Seasonality)

```python
ambj_adj_close_px=ambj['Adj Close']
#Short moving window rolling mean

ambj['42'] = ambj_adj_close_px.rolling(window=40).mean()

# Long moving window rolling mean
ambj['252'] = ambj_adj_close_px.rolling(window=252).mean()

# Plot the adjusted closing price, the short and long windows of rolling means
ambj[['Adj Close', '42', '252']].plot(title="AMBUJA CEMENTS")

# Show plot
plt.show()

knst_adj_close_px = knst['Adj Close']
# Short moving window rolling mean
knst['42'] = knst_adj_close_px.rolling(window=40).mean()

# Long moving window rolling mean
knst['252'] = knst_adj_close_px.rolling(window=252).mean()

# Plot the adjusted closing price, the short and long windows of rolling means
knst[['Adj Close', '42', '252']].plot(title="KUNSTOFF INDUSTRIES")

# Show plot
plt.show()

niti_adj_close_px = niti['Adj Close']
# Short moving window rolling mean
niti['42'] = niti_adj_close_px.rolling(window=40).mean()

# Long moving window rolling mean
niti['252'] = niti_adj_close_px.rolling(window=252).mean()

# Plot the adjusted closing price, the short and long windows of rolling means
niti[['Adj Close', '42', '252']].plot(title="NITIN SPINNERS")

# Show plot
plt.show()

hpct_adj_close_px = hpct['Adj Close']
# Short moving window rolling mean
hpct['42'] = hpct_adj_close_px.rolling(window=40).mean()

# Long moving window rolling mean
hpct['252'] = hpct_adj_close_px.rolling(window=252).mean()

# Plot the adjusted closing price, the short and long windows of rolling means
hpct[['Adj Close', '42', '252']].plot(title="HP COTTONS")

# Show plot
plt.show()
```
